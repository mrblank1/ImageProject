# mini project 1

یک مقدار کند است برای همین ویدیو خروجی قرار داده میشود و حجم مدل سنگین است نمیتوان لود کرد

```python
import cv2

# Paths for the pre-trained model files
protoFile = "pose/coco/pose_deploy_linevec.prototxt"
weightsFile = "pose/mpi/pose_iter_440000.caffemodel"

# Read the network into Memory
net = cv2.dnn.readNetFromCaffe(protoFile, weightsFile)

# Specify the input image dimensions
inWidth = 368
inHeight = 368

# Threshold to filter out keypoints with low probability
threshold = 0.1

# Function to estimate pose on each frame
def estimate_pose(frame):
    # Prepare the frame to be fed to the network
    inpBlob = cv2.dnn.blobFromImage(frame, 1.0 / 255, (inWidth, inHeight), (0, 0, 0), swapRB=False, crop=False)

    # Set the prepared object as the input blob of the network
    net.setInput(inpBlob)
    out = net.forward()
    H = out.shape[2]
    W = out.shape[3]

    # Empty list to store the detected keypoints
    points = []
    for i in range(out.shape[1]):
        # confidence map of corresponding body's part.
        probMap = out[0, i, :, :]

        # Find global maxima of the probMap.
        minVal, prob, minLoc, point = cv2.minMaxLoc(probMap)

        # Scale the point to fit on the original image
        x = (frame.shape[1] * point[0]) / W
        y = (frame.shape[0] * point[1]) / H

        if prob > threshold:
            # Reduce the circle radius to 8 and the font size to 0.6 for smaller visualizations
            cv2.circle(frame, (int(x), int(y)), 8, (0, 255, 255), thickness=-1, lineType=cv2.FILLED)
            cv2.putText(frame, "{}".format(i), (int(x), int(y)), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2,
                        lineType=cv2.LINE_AA)

            # Add the point to the list if the probability is greater than the threshold
            points.append((int(x), int(y)))
        else:
            points.append(None)

    return frame

# Open video capture (replace 'path/to/your/video.mp4' with the path to your video file)
input_file = 'pose.mp4'
cap = cv2.VideoCapture(input_file)

# Get the video frame dimensions
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# Define the codec and create VideoWriter object to save the output video
output_file = 'output_video.avi'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_file, fourcc, 30, (frame_width, frame_height))

while True:
    ret, frame = cap.read()

    if not ret:
        break

    # Call the function to estimate pose on the current frame
    frame_with_pose = estimate_pose(frame)

    # Write the frame with pose estimation to the output video
    out.write(frame_with_pose)

    cv2.imshow("Pose Estimation", frame_with_pose)

    if cv2.waitKey(1) & 0xFF == 27:  # Press 'Esc' to exit
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2

def detect_faces(image_path, face_cascade_path):
    # Load the Haar Cascade classifier for face detection
    face_cascade = cv2.CascadeClassifier(face_cascade_path)

    # Read the input image
    image = cv2.imread(image_path)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces in the grayscale image
    faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    return faces

def main():
    # Path to the pre-trained Haar Cascade classifier for face detection
    face_cascade_path = "haarcascade_frontalface_alt.xml"  # Replace with the path to the XML file

    # Load the images
    image1_path = "face.jpg"  # Replace with the path to your first image
    image2_path = "people.jpg"  # Replace with the path to your second image

    # Detect faces in the images
    faces1 = detect_faces(image1_path, face_cascade_path)
    faces2 = detect_faces(image2_path, face_cascade_path)

    # Load the images using OpenCV for visualization
    image1 = cv2.imread(image1_path)
    image2 = cv2.imread(image2_path)

    # Draw rectangles around the faces and label them
    for (x, y, w, h) in faces1:
        cv2.rectangle(image1, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(image1, "Person 1", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    for (x, y, w, h) in faces2:
        cv2.rectangle(image2, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(image2, "Person 2", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Show the result images
    cv2.imshow("Image 1", image1)
    cv2.imshow("Image 2", image2)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

# mini project 2

با استفاده از عکسی که در رفرنس داده میشود . اگر یکسان باشد با فرد داخل وب کم دور آن سبز در غیر اینصورت بقیه افراد قرمز هستند

```python
import cv2
import face_recognition

def compare_faces(reference_image_path, target_frame):
    # Load the reference image (face of the person to recognize)
    reference_image = face_recognition.load_image_file(reference_image_path)
    reference_encoding = face_recognition.face_encodings(reference_image)[0]

    # Find all face locations in the target frame
    target_face_locations = face_recognition.face_locations(target_frame)
    target_face_encodings = face_recognition.face_encodings(target_frame, target_face_locations)

    # Compare face encodings for matches
    face_matches = face_recognition.compare_faces(target_face_encodings, reference_encoding)

    return target_face_locations, face_matches

def main():
    # Path to the reference image (face of the person to recognize)
    reference_image_path = "face.jpg"  # Replace with the path to your reference image

    
    # Load the reference image using face_recognition for later comparison
    reference_image = face_recognition.load_image_file(reference_image_path)
    reference_encoding = face_recognition.face_encodings(reference_image)[0]

    # Open the video capture
    video_capture = cv2.VideoCapture(0)

    while True:
        # Read a frame from the video capture
        ret, frame = video_capture.read()

        if not ret:
            break

        # Compare faces in the frame with the reference face
        face_locations, face_matches = compare_faces(reference_image_path, frame)

        # Draw rectangles around the faces and label them
        for (top, right, bottom, left), match in zip(face_locations, face_matches):
            color = (0, 255, 0) if match else (0, 0, 255)
            cv2.rectangle(frame, (left, top), (right, bottom), color, 2)
            label = "Person" if match else "Unknown"
            cv2.putText(frame, label, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        # Display the result frame
        cv2.imshow("Video with Face Recognition", frame)

        # Break the loop when 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the video capture and close the display window
    video_capture.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

# متاسفانه در مینی پروژه ۳ زیاد موفق نبودم

```python
import cv2

# Choose the tracker method
tracker_type = 'CSRT'

# Initialize the selected tracker
if tracker_type == 'BOOSTING':
    tracker = cv2.TrackerBoosting_create()
elif tracker_type == 'MIL':
    tracker = cv2.TrackerMIL_create()
elif tracker_type == 'KCF':
    tracker = cv2.TrackerKCF_create()
elif tracker_type == 'CSRT':
    tracker = cv2.TrackerCSRT_create()
elif tracker_type == 'TLD':
    tracker = cv2.TrackerTLD_create()
elif tracker_type == 'MEDIANFLOW':
    tracker = cv2.TrackerMedianFlow_create()
elif tracker_type == 'GOTURN':
    tracker = cv2.TrackerGOTURN_create()
elif tracker_type == 'MOSSE':
    tracker = cv2.TrackerMOSSE_create()
else:
    print('Invalid tracker type.')
    exit()

# Open the video file (replace 'path_to_video' with the actual path to your video file)
video_path = 'race_car.mp4'
cap = cv2.VideoCapture(video_path)

# Read the first frame to select the object to track
ret, frame = cap.read()
if not ret:
    print('Error opening video file.')
    exit()

# Select the object to track (use the bounding box of the object)

bbox = cv2.selectROI('Select Object to Track', frame, False)
cv2.destroyAllWindows()

# Initialize the tracker with the selected object bounding box
bbox = tuple(map(int, bbox))  # Convert bounding box coordinates to integers
ok = tracker.init(frame, bbox)



while True:
    # Read a new frame
    ret, frame = cap.read()
    if not ret:
        break

    # Update the tracker
    ok, bbox = tracker.update(frame)

    # If the tracking is successful, draw the bounding box
    if ok:
        x, y, w, h = [int(v) for v in bbox]
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
    else:
        print('Tracking failure.')

    # Display the resulting frame
    cv2.imshow('Object Tracking', frame)

    # Check if the 'q' key is pressed to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
